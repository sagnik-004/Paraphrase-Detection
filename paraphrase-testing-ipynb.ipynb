```python
import os
import zipfile
from sentence_transformers import SentenceTransformer
import torch
from sentence_transformers.util import cos_sim
```

```python
model_path = "/kaggle/input/paraphrase-project-backup-fixed/paraphrase_project_backup/sbert-paraphrase-2025-04-16_16-28-38/saved_model_paraphrase_0416"
model = SentenceTransformer(model_path)

print("Model loaded successfully!")
```

```python
from sentence_transformers.util import cos_sim

def is_paraphrase(sentence1, sentence2, threshold=0.75):
    embeddings = model.encode([sentence1, sentence2], convert_to_tensor=True)
    similarity = cos_sim(embeddings[0], embeddings[1]).item()
    result = "Paraphrased" if similarity >= threshold else "Not Paraphrase"
    return result, similarity

# Test example
s1 = "He is going to school."
s2 = "He attends school."

result, score = is_paraphrase(s1, s2)
print("sentence1: ",s1)
print("sentence2: ",s2)
print(f"Cosine Similarity: {score:.4f} -----> {result}")
```
